```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import pickle
```

```python
with open('/content/drive/MyDrive/data (2)/data/mnist_small.pkl', 'rb') as file:
    data = pickle.load(file)
```

```python
X = data['X']  # Features
Y = data['Y']  # Labels
```

```python
pca = PCA(n_components=2)
pca_result = pca.fit_transform(X)
```

```python
plt.scatter(pca_result[:, 0], pca_result[:, 1], c=Y, cmap='tab10')
plt.title('PCA')
plt.colorbar(ticks=range(10))
plt.show()
```

```python
tsne = TSNE(n_components=2, random_state=42)
tsne_result = tsne.fit_transform(X)
```

```python
plt.figure("tsne_result")
plt.scatter(x=tsne_result[:,0],y=tsne_result[:,1],c=Y, cmap='tab10')
plt.title("tSNE")
plt.colorbar(ticks=range(10))
plt.show()
```
